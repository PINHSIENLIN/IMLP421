# 匯入在sklearn.linear_model套件裡面的LinearRegression模型
# 匯入在sklearn.model_selection套件裡面的train_test_split模組
# 切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
# 畫出訓練資料集的matplotlib圖形

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 20)
```

```python
reg_model = LinearRegression()
predict_model = reg_model.fit(x_train.reshape(-1,1), y_train)
```

## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
plt.scatter(x_train, y_train, label = "Training Data")
plt.plot(x_train, predict_model.predict(x_train.reshape(-1,1)), "r", label = "Predicted Model")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.square(y_train - predict_model.predict(x_train.reshape(-1,1))).sum() / len(y_train)
```

```python
mean_squared_error(y_train, predict_model.predict(x_train.reshape(-1,1)))
```
